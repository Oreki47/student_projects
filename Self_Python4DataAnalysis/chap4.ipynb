```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points = np.arange(-5, 5, 0.01)
xs, ys= np.meshgrid(points, points)
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)
```

```python
plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
plt.show()
```

```python
# np.where example
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
np.where(cond, xarr, yarr)
```

```python
# np.random example
nsteps = 1000
walk = np.random.randint(0, 2, size=1000)
walk = np.where(walk > 0, 1, -1)
walk = np.cumsum(walk)
plt.plot(walk)
plt.show()
```

```python
nwalks = 5000
nsteps = 1000
walks = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(walks > 0, 1, -1)
walks = steps.cumsum(1)
walks
```

```python
hist30 = (np.abs(walks) > 30).any(1)
hist30.sum()
```

```python
crossing_time = (np.abs(walks[hist30]) > 30).argmax(1)
```

```python
crossing_time.mean()
```
